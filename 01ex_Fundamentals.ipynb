2\. Convert the following function into a pure function with no global variables or side effects

````python
import copy

x = 5
def f(alist):
    for i in range(x):
         alist.append(i)
    return alist

alist = [1,2,3]
ans = f(alist)
print (ans)
print (alist) # alist has been changed!

#modified
#x = 5
def f(alist):
    new = copy.copy(alist) #shallow copy garantisce che non ci siano modifiche nella cosa copiata
    for i in range(5):
        new.append(i)
    return new

alist = [1,2,3]
ans = f(alist)
print (ans)
print (alist) # alist has been changed!
````

3\. Write a `decorator` hello that makes every wrapped function print “Hello!”, i.e. something like:

```python
@hello
def square(x):
    return x*x
```



````python
def hello (func):
    def wrapper (x):
        print("Hello")
        func(x)
        return func(x) #devo mettere il return alla funzione se la funzione ha un argomento. c'è un modo per generalizzare gli argomenti in modo che anche se uso lo stesso wrap per funzioni con numero di argomenti diversi funziona    
    return wrapper
    
@hello
def square(x):
    return x*x

print(square(5))
````

4\. Write the factorial function so that it a) does and b) does not use recursion.

````python
def factorial (x):
    if x == 0:
        return 1
    else: return x*factorial(x-1)

print(factorial(3))

def factorial_1(n):
   # """Fib without recursion."""
    f=1
    for i in range(1, n+1):
        f = i*f 
    return f
print(factorial_1(3))
````

5\. Use HOFs (zip in particular) to compute the weight of a circle, a disk and a sphere, assuming different radii and different densities:

```python
densities = {"Al":[0.5,1,2],"Fe":[3,4,5],"Pb": [15,20,30]}
radii = [1,2,3]
```

where the entries of the dictionary's values are the linear, superficial and volumetric densities of the materials respectively.

In particular define a list of three lambda functions using a comprehension that computes the circumference, the area and the volume for a given radius.


````python
import math

# Dati
densities = {"Al":[0.5,1,2], "Fe":[3,4,5], "Pb":[15,20,30]}
radii = [1,2,3]  # tutti i raggi da usare su tutti i materiali
materials = list(densities.keys())

# Lambda functions per geometria usando comprehension: La list comprehension [f for f in (...)] prende ciascun lambda e lo mette in una lista gni elemento della lista è una funzione che può essere chiamata con geom(r)
geometries = [f for f in (  
    lambda r: 2*math.pi*r,        # lineare → circonferenza
    lambda r: math.pi*r**2,       # superficiale → area
    lambda r: (4/3)*math.pi*r**3  # volumetrica → volume
)]

# Calcolo dei pesi: ogni raggio con tutti i materiali
weights_per_radius = [
    {
        mat: [density * geom(r) for density, geom in zip(densities[mat], geometries)]
        for mat in materials
    }
    for r in radii
]

# Stampa dei risultati
for i, r in enumerate(radii):
    print(f"Raggio = {r}")
    for mat, w in weights_per_radius[i].items():
        print(f"  {mat}: lineare={w[0]:.2f}, superficiale={w[1]:.2f}, volumetrica={w[2]:.2f}")
    print()
````

# Analisi dettagliata

## List comprehension esterna: for r in radii

* Itera su ogni raggio della lista radii

* Ogni elemento della lista risultante sarà un dizionario di pesi per quel raggio.

## Dictionary comprehension interna: for mat in materials

* Per ogni materiale (Al, Fe, Pb) creiamo una voce nel dizionario

    * Chiave = nome del materiale

    * Valore = lista dei pesi calcolati per tutte le densità del materiale

## Lista interna: [density * geom(r) for density, geom in zip(densities[mat], geometries)]

* densities[mat] → lista [lineare, superficiale, volumetrica]

* geometries → lista di lambda [circ, area, volume]

* zip(densities[mat], geometries) → allinea densità e funzione:

    * lineare ↔ circonferenza

    * superficiale ↔ area

    * volumetrica ↔ volume

* density * geom(r) → calcola il peso usando la densità e il raggio corrente r

* Risultato → lista [peso_lineare, peso_superficie, peso_volumetrico] per quel materiale e raggio

## Struttura finale di weights_per_radius

* Tipo: list[dict[str, list[float]]]

* Ogni elemento della lista → un raggio

* Ogni dizionario → tutti i materiali

* Lista interna → i tre tipi di peso [lineare, superficiale, volumetrico]


## Stampa dell'output

* enumerate(radii) → otteniamo indice i e raggio r

* weights_per_radius[i] → prende il dizionario dei pesi per quel raggio

* .items() → scorre i materiali e i loro pesi

* w[0], w[1], w[2] → peso lineare, superficiale e volumetrico

* :.2f → stampa con 2 cifre decimali
