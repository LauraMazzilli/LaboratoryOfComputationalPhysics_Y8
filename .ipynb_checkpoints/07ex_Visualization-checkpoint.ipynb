1\. **Kernel Density Estimate**

Produce a KDE for a given distribution (by hand, not using seaborn!):

* Fill a numpy array, x,  of len(N) (with N=O(100)) with a variable normally distributed, with a given mean a standard deviation
* Fill an histogram in pyplot taking properly care about the aesthetic
   * use a meaningful number of bins
   * set a proper y axis label
   * set proper value of y axis major ticks labels (e.g. you want to display only integer labels)
   * display the histograms as data points with errors (the error being the poisson uncertainty)
* for every element of x, create a gaussian with the mean corresponding the element value and std as a parameter that can be tuned. The std default value should be:
$$ 1.06 * x.std() * x.size ^{-\frac{1}{5.}} $$
you can use the scipy function `stats.norm()` for that.
* In a separate plot (to be placed beside the original histogram), plot all the gaussian functions so obtained
* Sum (with np.sum()) all the gaussian functions and normalize the result such that the integral matches the integral of the original histogram. For that you could use the `scipy.integrate.trapz()` method


````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats, integrate
from numpy import trapezoid

# PARAMETERS
N = 200
mean = 0
sigma = 1

# KDE bandwidth (std of each small Gaussian)
kde_sigma = 0.3


# GENERATE DATA
np.random.seed(42)
x = np.random.normal(mean, sigma, N)


# HISTOGRAM WITH POISSON ERRORS
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Histogram
bins = 20
counts, bin_edges = np.histogram(x, bins=bins)
bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])
errors = np.sqrt(counts)        # Poisson uncertainties

# Plot histogram as points with error bars
ax1.errorbar(
    bin_centers, counts, yerr=errors,
    fmt='o', color='black', capsize=3
)

#ax1.set_xlabel("x")
one_tick = fig.axes[0].yaxis.get_major_ticks()[0]
ax1.set_ylabel("Counts")
ax1.set_title("Histogram with Poisson errors")

# Set integer y-axis ticks
one_tick = fig.axes[0].yaxis.get_major_ticks()[0]
ymax = counts.max() + 2
ax1.set_yticks(np.arange(0, ymax+1, 1))


# KDE BY SUM OF GAUSSIANS
# Fine grid for KDE
xx = np.linspace(x.min() - 1, x.max() + 1, 1000)

# Each Gaussian: store in a list for plotting
gaussians = []
for val in x:
    g = stats.norm.pdf(xx, loc=val, scale=kde_sigma)
    gaussians.append(g)
    ax2.plot(xx, g, alpha=0.2)   # plot each component

ax2.set_title("All Gaussian components")
ax2.set_xlabel("x")

# SUM THE GAUSSIANS AND NORMALIZE
kde_raw = np.sum(gaussians, axis=0)

# Normalize such that integral matches histogram total counts
target_area = np.sum(counts)
current_area = trapezoid(kde_raw, xx)
kde = kde_raw * (target_area / current_area)

# Plot the normalized KDE on the histogram for comparison
ax1.plot(xx, kde, color='red', linewidth=2, label="KDE (sum of Gaussians)")
ax1.legend()

plt.tight_layout()
plt.show()
````

2\. **Color-coded scatter plot**

Produce a scatter plot out of a dataset with two categories

* Write a function that generate a 2D datasets of 2 categories. Each category should distribute as a 2D gaussian with a given mean and std (clearly it is better to have different values means..)
* Display the dataset in a scatter plot marking the two categories with different marker colors.

An example is given below

You can try to make the procedure more general by allowing a given number $n\ge 2$ of categories

````python
import numpy as np
import matplotlib.pyplot as plt

# FUNCTION: generate a 2D dataset with k categories
def generate_2d_gaussian_dataset(k=2, n_per_class=200):
    """
    Generates a 2D dataset with k categories.
    Each category is a 2D Gaussian with its own mean and covariance.
    """
    data = []
    labels = []

    for i in range(k):
        # Random mean for each category
        mean = np.random.uniform(-5, 5, size=2)

        # Random covariance matrix (positive definite)
        A = np.random.rand(2, 2)
        cov = np.dot(A, A.T)  # ensures covariance matrix is positive definite

        # Generate gaussian cluster
        cluster = np.random.multivariate_normal(mean, cov, n_per_class)

        data.append(cluster)
        labels.append(np.full(n_per_class, i))

    # Combine all clusters
    data = np.vstack(data)
    labels = np.concatenate(labels)

    return data, labels


# GENERATE DATA
data, labels = generate_2d_gaussian_dataset(k=2, n_per_class=300)


# PLOT
plt.figure(figsize=(7, 6))

# Use a color per category
for cat in np.unique(labels):
    plt.scatter(
        data[labels == cat, 0],
        data[labels == cat, 1],
        s=30,
        label=f"Category {cat}"
    )

plt.xlabel("x")
plt.ylabel("y")
plt.title("2D Gaussian Categories")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
````
