2\. Create a random matrix of 5x5 and then save it to a text file named "data.txt"

```python
m = np.random.randint(0,25,25)
n = m.reshape(5,5)
print(n)

with open('C:/Users/mazzi/OneDrive/Desktop/PoD/data.txt', 'w') as f: #da qui in poi doppio ciclo tipico per scorrere sulle matrici
    for l in range(5):
        for m in range(5):
            f.write(str(n[l][m]) + '\t')
        f.write ('\n')
```

3\. Load the saved txt file of point 2 and convert it to a csv file (by hand)

```python
with open('C:/Users/mazzi/OneDrive/Desktop/PoD/data.txt', mode='r') as f:
    with open('C:/Users/mazzi/OneDrive/Desktop/PoD/data.csv', mode='w') as f1:
        for line in f:
            index = 0
            for c in line.split():
                if index != 0: #questo serme per togliere la virgola dalla prima colonna
                    f1.write(',' + c)
                else: f1.write(c) 
                index +=1
            f1.write('\n')
```
