````python
# 1
ans = []
for i in range(3):
    for j in range(4):
        ans.append((i, j))
print (ans)

ans_n = [(i,j) for i in range(3) for j in range(4)]
print("io", ans_n)

# 2
ans = map(lambda x: x*x, filter(lambda x: x%2 == 0, range(5))) #ans è tipo un puntatore
print (list(ans)) #evidentemente ans non è una lista
print(ans)

ans_n = [i*i for i in range(5) if i%2 == 0]
print("io", ans_n)
````

2\. Convert the following function into a pure function with no global variables or side effects

````python
import copy

x = 5
def f(alist):
    for i in range(x):
         alist.append(i)
    return alist

alist = [1,2,3]
ans = f(alist)
print (ans)
print (alist) # alist has been changed!

#modified
#x = 5
def f(alist):
    new = copy.copy(alist) #shallow copy garantisce che non ci siano modifiche nella cosa copiata
    for i in range(5):
        new.append(i)
    return new

alist = [1,2,3]
ans = f(alist)
print (ans)
print (alist) # alist has been changed!
````

3\. Write a `decorator` hello that makes every wrapped function print “Hello!”, i.e. something like:

```python
@hello
def square(x):
    return x*x
```



````python
def hello (func):
    def wrapper (x):
        print("Hello")
        func(x)
        return func(x) #devo mettere il return alla funzione se la funzione ha un argomento. c'è un modo per generalizzare gli argomenti in modo che anche se uso lo stesso wrap per funzioni con numero di argomenti diversi funziona    
    return wrapper
    
@hello
def square(x):
    return x*x

print(square(5))
````

4\. Write the factorial function so that it a) does and b) does not use recursion.

````python
def factorial (x):
    if x == 0:
        return 1
    else: return x*factorial(x-1)

print(factorial(3))

def factorial_1(n):
   # """Fib without recursion."""
    f=1
    for i in range(1, n+1):
        f = i*f 
    return f
print(factorial_1(3))
````
